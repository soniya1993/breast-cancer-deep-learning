```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#loading the data with sklearn
df_breastcancer=sns.datasets.load_breast_cancer()
```

```python
df_breastcancer
```

```python
df = pd.DataFrame(df_breastcancer.data, columns = df_breastcancer.feature_names)
```

```python
df.head()
```

```python
df['label']=df_breastcancer.target
```

1 --> Benign

Separating the features and target

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
#Splitting the data into training data & Testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

Standardize the data

```python
#importing library for standarization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

Building the neural network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network
model=keras.Sequential([
                        keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(2,activation='sigmoid')


])
```

```python
# compiling the Neural Network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Meural Network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_accuracy'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
Y_pred = model.predict(X_test_std)
```

```python
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```
